#Variables

```python
# Variables and Data Types

# Integer
x = 10
print(f"x (integer): {x}")

# Float
y = 3.14
print(f"y (float): {y}")

# String
name = "Python"
print(f"name (string): {name}")

# Boolean
is_coding = True
print(f"is_coding (boolean): {is_coding}")
```

#Basic Operation

```python
# Basic Operations

# Addition
a = 5
b = 3
sum_ab = a + b
print(f"Sum of {a} and {b}: {sum_ab}")

# Subtraction
difference_ab = a - b
print(f"Difference of {a} and {b}: {difference_ab}")

# Multiplication
product_ab = a * b
print(f"Product of {a} and {b}: {product_ab}")

# Division
quotient_ab = a / b
print(f"Quotient of {a} and {b}: {quotient_ab}")

# Modulus
remainder_ab = a % b
print(f"Remainder of {a} divided by {b}: {remainder_ab}")

# Exponentiation
power_ab = a ** b
print(f"{a} raised to the power of {b}: {power_ab}")
```

#Control Flow (If else - Loop)

```python
# If-Else Statements
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")

# Elif Statements
y = 7
if y > 10:
    print("y is greater than 10")
elif y == 7:
    print("y is 7")
else:
    print("y is less than 7")
```

```python
# Loops
#i 10; condition; inc++/
# For loop
print("For loop:")
for i in range(2,10,2):
    print(f"i: {i}")

# While loop
print("While loop:")
count = 0
while count < 5:
    print(f"count: {count}")
    count += 1
```

#Function

```python
#Functions

def greet(a):
    return f"Hello, {a}!"

# Call the function
name1 = greet("Alice")
print(name1)

# Function with multiple parameters
def add(a, b):
    return a + b

# Call the function
res = add(10,5)
print(res)
```

#Lists and Dictionarys

```python
#List

# Create a list
fruits = ["apple", "banana", "cherry"]
print(fruits)

# Access list elements
print(fruits[0])

# Append to a list
fruits.append("date")
print(fruits)

```

```python
# Dictionaries

# Create a dictionary
student = {"name": "John", "age": 21, "courses": ["Math", "Science"], 'k':(1,2,3)}
print(student)

# Access dictionary values
print(student['name'])

# Add a new key-value pair
student["grade"] = "A"
print(student)

print(student["courses"][0])
```

```python
#Tuple Operation
tuple1 = (1, 2)
tuple2 = (3, 4)
tuple3 = (6, 7)

combined = tuple1 + tuple2  + tuple3 # (1, 2, 3, 4)
print(combined)
```

```python
#Repetition: Repeating elements in a tuple using the * operator:
repeated = (1, 2) * 3         # (1, 2, 1, 2, 1, 2)

print(repeated)
```

```python
#Membership Test:
t1 = (1, 2, 3)
exists = 5 in t1    # True
print(exists)
```

Simple grading system using python

```python


def calculate_grade(score):
  if score >= 80:
    return "A+"
  elif score >= 75:
    return "A"
  elif score >= 70:
    return "B+"
  elif score >= 65:
    return "B"
  elif score >= 60:
    return "C+"
  elif score >= 55:
    return "C"
  else:
    return "F"

score = 78
grade = calculate_grade(score)
print(f"Your score is {score} and your grade is {grade}")
```

#task 02
create a code that can generate n amount of values for the following series . n=1,5,3,7,5..........

```python

import numpy as np

def generate_pattern(n):

  pattern = []
  for i in range(n):
    if i % 2 == 0:
      pattern.append(i // 2 * 2 + 1)
    else:
      pattern.append(i // 2 * 2 + 5)
  return pattern

n = 10
pattern = generate_pattern(n)
print(pattern)
```
